## Librerias

```python
# Importamos las librerias que usaremos para el programa
import numpy as np
import random as rnd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.integrate import quad
```

## Método

```python
# Definimos una función para calcular Método de la Composición
def Composicion(landa, n):
    L = 1 / landa
    lista = []
    for i in range(n):
        U = rnd.uniform(0, 1) # Generamos números aleatorios distribuidos uniformemente de 0 a 1
        V = rnd.uniform(0, 1)
        if U < 0.5:
            X = -L * np.log(1 - V)
        else:
            X = L * np.log(V)
        lista.append(X)
    return lista

Resultados = Composicion(1.5, 10000)
```

## Función de densidad de probabilidad

```python
# Calcula la Función de densidad de probabilidad para diferentes valores de x
def fdp(X, landa):
    L = 1 / landa
    if X > 0:
        return 0.5 * np.exp(- X / L) / L
    elif X < 0:
        return 0.5 * np.exp(X / L) / L
    else:
        return 0

landa = 1.5 # Tomamos a 1.5 como valor de lambda
x = np.linspace(-10, 10, 100)
y = [fdp(x, landa) for x in x]

# Grafica de la Función de densidad de probabilidad
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Función de densidad de probabilidad')
plt.show()
```

## Función acumulativa

```python
# Calcula la Función Acumulativa para diferentes valores de x
def fda(X, landa):
    integral, _ = quad(fdp, -np.inf, X, args=(landa,))
    return integral

landa = 1.5
x = np.linspace(-10, 10, 100)
y = [fda(x, landa) for x in x]

# Grafica de la Función Acumulativa
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('Función Acumulativa')
plt.show()
```

## Histograma de la muestra
